# ***Tratando os dados***

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

# ***Matplotlib***

```python
import matplotlib.pyplot as plt
```

## Plotando gráficos

```python
plt.figure(figsize=(8,4))
# altera o tamanho do gráfico: figsize(largura, altura)

plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
# plt.plot(eixo x, eixo y)

plt.title('Imigração do Brasil para o Canadá')
# adiciona o título do gráfico

plt.xlabel('Ano')
# rótulo do eixo x

plt.ylabel('Número de imigrantes')
# rótulo do eixo y

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# escolhendo quais anos vão aparecer no eixo x

plt.show()
# exibe apenas o gráfico
```

## Criando uma figura

```python
fig, ax = plt.subplots(figsize=(8,4))
# fig -> cria uma figura
# ax -> modificar algumas coisas do eixos

ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
# cria um gráfico com os valores de ano no eixo x, e qtdd de imigrantes no eixo y

ax.set_title('Imigração do Brasil para o Canadá\n1980-2013')
# define o título do gráfico

ax.set_xlabel('Ano')
# define o rótulo do eixo x

ax.set_ylabel('Número de imigrantes')
# define o rótulo do eixo y

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# posiciona os marcadores em intervalos de 5 unidades

plt.show()
```

## Criando subplots em uma direção

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))
# cria uma figura com duas subplots, com uma linha e duas colunas(1, 2)

# define as características do primeiro subplot (gráfico de linha)
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

# define a caractorísticas do segundo subplot (boxplot)
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()

plt.show()
```

## Criando subplots em duas direções

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos quatro maiores países da América do sul para o Canadá de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

## Personalizando visualizações

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3)
# lw modifica a espessura da linha

ax.set_title('Imigração do Brasil para o Canadá\n1988 a 2013', fontsize=18, loc='left')
# fontsize antera o tamanho da fonte
# loc posiciona o título

ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)

# definimos o tamanho dos ticks do eixo x e y
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# adiciona uma grade pontilhada no fundo do gráfico
plt.grid(linestyle='--')

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
# Cria uma lista vazia para armazenar as cores que serão usadas no gráfico de barras.
cores = []
# Itera sobre os índices (nomes dos países) de america_sul_sorted.
for pais in america_sul_sorted.index:
    # Se o país for 'Brasil', adiciona 'green' (verde) à lista de cores.
    if pais == 'Brasil':
        cores.append('green')
    # Se o país não for 'Brasil', adiciona 'silver' (prata) à lista de cores.
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))

# Cria um gráfico de barras horizontais (barh) usando os índices (nomes dos países) de america_sul_sorted
# e o total de imigrantes como valores, aplicando a lista de cores criada anteriormente.
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)

ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013',
             loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)

# Remove o rótulo do eixo Y, deixando-o vazio.
ax.set_ylabel('')

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

Adicionando anotações

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Itera sobre os valores e índices da coluna 'Total' de america_sul_sorted.
# 'i' representa o índice da iteração, e 'v' representa o valor correspondente na coluna 'Total'.
for i, v in enumerate(america_sul_sorted['Total']):
    # Adiciona texto ao gráfico de barras horizontais.
    # O texto é o valor 'v' convertido em string, posicionado no eixo X em v + 20 (20 unidades à direita da barra).
    # O texto é posicionado no eixo Y em 'i' (a posição da barra correspondente).
    # O texto é preto ('black'), com tamanho de fonte 10, alinhado à esquerda (ha='left') e centralizado verticalmente (va='center').
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

# Remove o quadro (a borda) ao redor do gráfico.
ax.set_frame_on(False)

# Oculta o eixo X, ou seja, a linha do eixo e os rótulos dos ticks (valores) do eixo X não serão exibidos.
ax.get_xaxis().set_visible(False)

# Remove as marcações (ticks) dos eixos X e Y.
# O argumento 'length=0' define que as marcações não terão comprimento, tornando-as invisíveis.
ax.tick_params(axis='both', which='both', length=0)


plt.show()
```

Personalizando visualizações

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Top 10 países com mais imigrantes para o Canadá\n1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('', fontsize=14)

plt.show()
```

Explorando paletas

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
  ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
  ax.set_xlabel('Número de imigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
# paletas de cores sequenciais (quanto mais escura a cor maior o valor)
gerar_grafico_paleta('Blues')
gerar_grafico_paleta('Blues_r')
```

```python
# paleta de cores divergentes
gerar_grafico_paleta('rocket')
```

```python
# para dados que variam em torno de um valor central
gerar_grafico_paleta('coolwarm')
```

```python
# paletas categóricas
gerar_grafico_paleta('tab10')
```

Explorando temas

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```

Retirando as linhas de cima e da direita

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette='tab10')

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
# essa é a função que retira as linhas
sns.despine()

plt.show()
```
